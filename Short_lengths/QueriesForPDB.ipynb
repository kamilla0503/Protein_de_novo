```python
from rcsbsearch import TextQuery
from rcsbsearch import rcsb_attributes as attrs
from rcsbsearch import Attr, TextQuery
```

```python
SEQUENCE_LENGTH_MAX = 80 
RESOLUTION_MAX = 2
CELL_LENGTH = 30 
```

```python
q1 = attrs.entity_poly.rcsb_entity_polymer_type == "Protein"
q2 = attrs.rcsb_entry_info.diffrn_resolution_high.value <= RESOLUTION_MAX
q3 = attrs.entity_poly.rcsb_sample_sequence_length <= SEQUENCE_LENGTH_MAX
q4 = attrs.rcsb_entry_info.deposited_polymer_entity_instance_count == 1
q5 = attrs.cell.length_a <= CELL_LENGTH 
q6 = attrs.cell.length_b <= CELL_LENGTH 
q7 = attrs.cell.length_c <= CELL_LENGTH 

query = q1 & q2 & q3 & q4 & q5 & q6 & q7


f = open('proteins_id.txt', 'w')

for assemblyid in query("entry"):
    #print(assemblyid)
    f.write(assemblyid)
    f.write(",")
    
f.close()
    
#for assemblyid in query("assembly"):
#    print(assemblyid)
```
